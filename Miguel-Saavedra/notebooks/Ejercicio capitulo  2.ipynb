```python
import mglearn
import matplotlib.pyplot as plt
```

```python
mglearn.discrete_scatter(X[:, 0], X[:, 1], y)
plt.legend(["Clase 0", "Clase 1"], loc=4)
plt.xlabel("Primera característica")
plt.ylabel("Segunda característica")
print("Forma de X: {}".format(X.shape))

```
